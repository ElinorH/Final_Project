```python
url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'
vacc_df = pd.read_csv(url)
```

```python
vacc_df['daily_vaccinations'].fillna(0 , inplace = True)
vacc_df
```

###### בשביל ה-״חיסון יומי״ החלטנו לשנות את הערכים החסרים לערך אפס כי הגיוני שלא יהיו מחוסנים נוספים כל יום.



```python
vacc_df['people_vaccinated_per_hundred']=vacc_df.groupby('location')[['people_vaccinated_per_hundred']].apply(lambda x: x.fillna(method='ffill'))
vacc_df
```
